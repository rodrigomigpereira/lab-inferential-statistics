```python
import scipy.stats as stats
import math

import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
sample_mean = 130.1
population_mean = 120
std_dev = 21.21
n = 100
```

```python
t_statistic = (sample_mean - population_mean) / (std_dev / math.sqrt(n))
```

```python
df = n - 1
```

```python
alpha = 0.05
critical_value = stats.t.ppf(1 - alpha/2, df)  
```

```python
if abs(t_statistic) > critical_value:
    print("Null hypothesis rejected: The sample mean is significantly different from the population mean.")
else:
    print("Null hypothesis not rejected: The sample mean is not significantly different from population mean")
```

```python
## Before we apply PCA we need to standardize/scale the data. For this we can use sklearn's preprocessing library 

from sklearn.preprocessing import StandardScaler
transformer = StandardScaler().fit(numerics)
scaled = transformer.transform(numerics)
scaled = pd.DataFrame(scaled)
scaled.columns = numerics.columns 
scaled.head()
```

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing
pca = PCA()
pca.fit(scaled)
```
